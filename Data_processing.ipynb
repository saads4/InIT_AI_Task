```python
trans['hour'] = trans['timestamp'].dt.hour
trans['day_of_week'] = trans['timestamp'].dt.day_name()
trans['month'] = trans['timestamp'].dt.month_name()
trans.head()
```

Extracting the following Hour of day, Day of week, Month 

```python
cust['email'].fillna(cust['email'].mode()[0], inplace=True)
prod['category'].fillna(prod['category'].mode()[0], inplace=True)
trans['payment_method'].fillna(trans['payment_method'].mode()[0], inplace=True)
```

replacing the empty catagorical value with mode

```python
cust['age'].fillna(cust['age'].median(), inplace=True)
trans['price'].fillna(trans['price'].median(), inplace=True)
trans['quantity'].fillna(trans['quantity'].median(), inplace=True)
```

replacing the empty numerical value with median

```python
trans['revenue'] = trans['price'] * trans['quantity']
```

Revenue calculation

```python
full_data = trans.merge(cust, on='customer_id', how='left')
```

```python
full_data = full_data.merge(prod, on='product_id', how='left')
```

```python
full_data.head(3)
```

merging the the data frame

```python
full_data['profit_margin'] = (full_data['price'] - full_data['cost_price'])/full_data['price']*100
```

calculation of profit margin

```python
full_data.head()
```

New Data Frame 

```python
cust_revenue = full_data.groupby('customer_id')['revenue'].sum().reset_index()
cust_revenue.head()
```

```python
cust_trans_count = full_data.groupby('customer_id')['transaction_id'].nunique().reset_index()
cust_trans_count.head()
```

```python
cust_avg_value = full_data.groupby('customer_id')['revenue'].mean().reset_index()
cust_avg_value.head()
```

```python
counts = full_data.groupby(['customer_id', 'category']).size().reset_index(name='prch_count')
top_idx = counts.groupby('customer_id')['prch_count'].idxmax()
cust_top_category = counts.loc[top_idx, ['customer_id', 'category']].rename(columns={'category': 'most_frequent_category'})
cust_top_category.head()
```

```python
cust_revenue.rename(columns={'revenue': 'total_revenue'}, inplace=True)
cust_trans_count.rename(columns={'transaction_id': 'No_trans'}, inplace=True)
cust_avg_value.rename(columns={'revenue': 'avg_trans_value'}, inplace=True)
```

```python
customer_summary = (
    cust_revenue
    .merge(cust_trans_count, on='customer_id', how='left')
    .merge(cust_avg_value, on='customer_id', how='left')
    .merge(cust_top_category, on='customer_id', how='left')
)
customer_summary.head()
```

```python
full_data['month_period'] = full_data['timestamp'].dt.to_period('M')
```

```python
monthly_summary = (
    full_data.groupby('month_period')
    .agg(
        total_revenue=('revenue', 'sum'),
        unique_customers=('customer_id', 'nunique'),
        avg_order_value=('revenue', 'mean')
    )
    .reset_index()
)
monthly_summary.head()
```

```python
monthly_summary['month_period'] = monthly_summary['month_period'].astype(str)
monthly_summary['month'] = pd.to_datetime(monthly_summary['month_period'])
monthly_summary = monthly_summary.sort_values('month').reset_index(drop=True)
monthly_summary.head()
```

```python
monthly_summary['mom_growth_%'] = monthly_summary['total_revenue'].pct_change() * 100
monthly_summary.head()
```

```python
product_summary = (
    full_data.groupby(['product_id', 'product_name', 'category'])
    .agg(
        total_revenue=('revenue', 'sum'),
        total_quantity=('quantity', 'sum'),
        avg_profit_margin=('profit_margin', 'mean')
    )
    .reset_index()
)
product_summary.head()
```

```python
top_by_revenue = product_summary.nlargest(10, 'total_revenue')
display(top_by_revenue)
```

```python
top_by_quantity = product_summary.nlargest(10, 'total_quantity')
display(top_by_quantity)
```

```python
top_by_profit_margin = product_summary.nlargest(10, 'avg_profit_margin')
display(top_by_profit_margin)
```
